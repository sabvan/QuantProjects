```python
import math
import random 
import numpy as np
import time
```

```python
class MentalMath:
    def __init__(self):
       
        self.problems = []
        
    def clearProblems(self):
        self.problems = []
         
    def getSettings(self):
        addMin, addMax = input("Addition and Subtraction Range: ").split()
        if int(addMax) != 0:
            self.addition(addMin, addMax)
         
        multMin, multMax = input("Multiplication and Division Range: ").split()
        if int(multMax) != 0:
            self.mult(multMin, multMax)
                  
        square = input("Squares Up To and Including: ")
        if int(square) != 0:
            self.squares(square)
        
        cubes = input("Cubes Up To and Including: ")
        if int(cubes) != 0:
            self.cubes(cubes)
        
        decMin, decMax = input("Decimals Add and Mult Range: ").split()
        if int(decMax) != 0:
            self.decimals(decMin, decMax)
        
        powerTwo = input("Powers of 2, 3, and 5 Up To and Including: ")
        if int(powerTwo) != 0:
            self.powerTwo(powerTwo)
  
        '''
        fracDenom, fracMax = input("Fraction Denoms: ").split()
        if int(fracDenom) != 0: 
            self.fractionDenom(fracDenom, fracMax)
        '''
        
        
    def play(self):
        endTime = time.time() + 60 * 10
        count = 0
        while time.time() < endTime:
            rand = random.randint(0, len(self.problems) - 1)
            temp = self.problems[rand]
            key, val = random.choice(list(temp.items()))
            
            inp = -1
            if rand == 4 or rand == 5:
                inp = float(input(key))
                while inp != val:
                    inp = float(input(key))
                
                print("Correct")
                count += 1
                
                
            else: 
                inp = int(input(key))
                while inp != val:
                    inp = int(input(key))
                
                print("Correct")
                count += 1
            
            
        print("Score: " + str(count))
            
        
        
    def addition(self, addMin, addMax):
        addProblems = {}
        for i in range(int(addMin), int(addMax) + 1):
            for j in range(i, int(addMax) + 1):
                sum = i + j
                addProblems[str(i) + " + " + str(j) + " = "] = sum
                addProblems[str(sum) + " - " + str(i) + " = "] = j
        self.problems.append(addProblems)
                
        
    def mult(self, multMin, multMax):
        multProblems = {}
        for i in range(int(multMin), int(multMax) + 1):
            for j in range(i, int(multMax) + 1):
                product = i * j
                multProblems[str(i) + " * " + str(j) + " = "] = product
                multProblems[str(product) + " % " + str(i) + " = "] = j
        self.problems.append(multProblems)
    
        
    def squares(self, square):
        squaresProblems = {}
        for j in range(int(square) + 1):
                squared = j * j
                squaresProblems[str(j) + "^2 = "] = squared
                squaresProblems["sqrt(" +  str(squared)  + ") = "] = j
        self.problems.append(squaresProblems) 
    
                
    def cubes(self, cubes):
        cubesProblems = {}
        for j in range(int(cubes) + 1):
                cubed = j ** 3
                cubesProblems[str(j) + "^3 = "] = cubed
                cubesProblems[ "cubeRoot(" +  str(cubed)  + ") = "] = j
                
        self.problems.append(cubesProblems) 
        
        
        
    def decimals(self, minDec, maxDec):
        decProblems = {}
        for i in np.arange(int(minDec), int(maxDec), 0.01):
            for j in np.arange(i, int(maxDec), 0.01):
                i = round(i, 2)
                j = round(j, 2)
                sum = round(i + j, 2)
                decProblems[str(i) + " + " + str(j) + " = "] = sum
                decProblems[str(sum) + " - " + str(i) + " = "] = j
    
        self.problems.append(decProblems)
        
        multDecProblems = {}
        for i in np.arange(int(minDec), int(maxDec), 0.01):
            for j in np.arange(i, int(maxDec), 0.1):
                i = round(i, 1)
                j = round(j, 1)
                product = round(i * j, 2)
                multDecProblems[str(i) + " * " + str(j) + " = "] = product
                multDecProblems[str(product) + " % " + str(i) + " = "] = j
                
        self.problems.append(multDecProblems)
        
        
        
    def powerTwo(self, powerT):
        powerProblems = {}
        for j in range(int(powerT) + 1):
            powered2 = 2 ** j
            powered3 = 3 ** j
            powered5 = 5 ** j
            
            powerProblems["2^(" + str(j) + ") = "] = powered2
            powerProblems[ "log_2(" +  str(powered2)  + ") = "] = j
            
            powerProblems["3^(" + str(j) + ") = "] = powered3
            powerProblems[ "log_3(" +  str(powered3)  + ") = "] = j
            
            powerProblems["5^(" + str(j) + ") = "] = powered5
            powerProblems[ "log_5(" +  str(powered5)  + ") = "] = j
            
        self.problems.append(powerProblems)
        
        
```

```python
temp = MentalMath()
temp.getSettings()
```

```python
temp.play()
```
